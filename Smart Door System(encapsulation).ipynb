```python
class Staff:
    def __init__(self, s_name, initial_code):
        self.s_name = s_name
        self.access_code = initial_code

    @property
    def access_code(self):
        return self.__access_code

    @access_code.setter
    def access_code(self, new_code):
        if len(new_code) == 6:
            self.__access_code = new_code
        else:
            print("Invalid Code!")


    def display_staff_info(self):
        print(self.s_name, self.access_code)


staff = Staff("Mr. Cobbinah", "UMAT07")
staff.display_staff_info()

staff.access_code = "423"
staff.access_code = "UMAT81"
staff.display_staff_info()
```
